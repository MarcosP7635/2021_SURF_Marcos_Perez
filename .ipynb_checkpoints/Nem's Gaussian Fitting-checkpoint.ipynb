```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Tue Aug 21 11:09:23 2018

@author: nem
"""


import numpy as np
import pyfits as py
import scipy as sp
from scipy import misc
from scipy.optimize import curve_fit
from scipy import signal
from math import *
import math as mt
from matplotlib.pyplot import *
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator
import sys
import os
import pdb
#import hot_pixels as hp
import pyfits as py
import matplotlib.colors as colors
interactive(True)

rcParams['image.origin'] = 'lower'         #Flips data about the horizontal axis 
rcParams['image.interpolation'] = 'nearest'# does not interpolate pixels in imshow

home = os.getenv('HOME')
#sys.path.append(home+'/src/lib/python/')
#from image_registration import chi2_shifts as CC

def center(im):
    Linemean1 = np.mean(im,axis=0)
    Linemean2 = np.mean(im,axis=1)
    ysize,xsize=np.shape(im)
    x = np.linspace(0,xsize-1,xsize)
    y = np.linspace(0,ysize-1,ysize)
    xcoor = np.argmax(Linemean1)                
    ycoor = np.argmax(Linemean2)
    return ycoor, xcoor
'''
def sqrt(im):
    scale_min = np.min(im)
    scale_max = np.max(im)
    im.clip(scale_min, scale_max)
    im = im - scale_min
    indices = np.where(im < 0)
    im[indices] = 0.0
    im = np.sqrt(im)
    im = im / mt.sqrt(scale_max - scale_min)
    return im
'''


#------------------------------------------------------------------#
#___________________________Import data____________________________#
wder    = '/Users/nem/Desktop/Bundle_2/'


data1   = py.getdata(wder+'Fiber_0_data_kband.fits')
data2   = py.getdata(wder+'Fiber_1_data_kband.fits')
data3   = py.getdata(wder+'Fiber_2_data_kband_optimum_focus.fits')
data4   = py.getdata(wder+'Fiber_3_data_kband.fits')
data5   = py.getdata(wder+'Fiber_4_data_kband.fits')
data6   = py.getdata(wder+'Fiber_5_data_kband.fits')
data7   = py.getdata(wder+'Fiber_6_data_kband.fits')
#data8   = py.getdata(wder+'-20micron_defocus_kband.fits')



datak   = np.array((data1, data2, data3, data4, data5, data6, data7))#, data8))
avedatak= np.mean(datak,axis=1) #averages each cube, but keeps 14 of them in the cube. 

#zsize,ysize,xsize = np.shape(datak) #Get dimensions of the input image cube

dark1   = py.getdata(wder+'Fiber_0_dark_kband.fits')
dark2   = py.getdata(wder+'Fiber_1_dark_kband.fits')
dark3   = py.getdata(wder+'Fiber_2_dark_kband.fits')
dark4   = py.getdata(wder+'Fiber_3_dark_kband.fits')
dark5   = py.getdata(wder+'Fiber_4_dark_kband.fits')
dark6   = py.getdata(wder+'Fiber_5_dark_kband.fits')
dark7   = py.getdata(wder+'Fiber_6_dark_kband.fits')

darkk   = np.array((dark1, dark2, dark3, dark4, dark5, dark6, dark7))#, dark8))
avedarkk = np.mean(darkk,axis=1) #This is an average dark frame

finalk   = avedatak - avedarkk


data12   = py.getdata(wder+'Fiber_0_data_Lband.fits')
data22   = py.getdata(wder+'Fiber_1_data_Lband.fits')
data32   = py.getdata(wder+'Fiber_2_data_Lband.fits')
data42   = py.getdata(wder+'Fiber_3_data_Lband_optimum_focus.fits')
data52   = py.getdata(wder+'Fiber_4_data_Lband.fits')
data62   = py.getdata(wder+'Fiber_5_data_Lband.fits')
data72   = py.getdata(wder+'Fiber_6_data_Lband.fits')
#data8   = py.getdata(wder+'-20micron_defocus_kband.fits')

dataL   = np.array((data12, data22, data32, data42, data52, data62, data72))#, data8))
avedataL= np.mean(dataL,axis=1) #averages each cube, but keeps 14 of them in the cube. 

#zsize,ysize,xsize = np.shape(datak) #Get dimensions of the input image cube

dark12   = py.getdata(wder+'Fiber_0_dark_Lband.fits')
dark22   = py.getdata(wder+'Fiber_1_dark_Lband.fits')
dark32   = py.getdata(wder+'Fiber_2_dark_Lband.fits')
dark42   = py.getdata(wder+'Fiber_3_dark_Lband.fits')
dark52   = py.getdata(wder+'Fiber_4_dark_Lband.fits')
dark62   = py.getdata(wder+'Fiber_5_dark_Lband.fits')
dark72   = py.getdata(wder+'Fiber_6_dark_Lband.fits')

darkL    = np.array((dark12, dark22, dark32, dark42, dark52, dark62, dark72))#, dark8))
avedarkL = np.mean(darkL,axis=1) #This is an average dark frame

finalL   = avedataL - avedarkL

#-----------------------Cropping data----------------------------#
yck     = 125
ycL     = 130
window  = 25

ImCropk = finalk[:,yck-window:yck+window,:]
ImCropL = finalL[:,ycL-window:ycL+window,:]
#yc, xc  = center(final)



#-----------------------Plotting data----------------------------#
plt.figure()

plt.imshow(ImCropk[2], cmap='viridis')
plt.ylim(0,50)
plt.ylabel('Pixel number')
plt.xlabel('Pixel number')
plt.show()
#fig1.savefig('Defocus_of_bundle1_output_kband_f=300mm', bbox_inches='tight', dpi=300)

plt.figure()

plt.imshow(ImCropL[3], cmap='viridis')
plt.ylim(0,50)
plt.ylabel('Pixel number')
plt.xlabel('Pixel number')
plt.show()
#fig1.savefig('Defocus_of_bundle1_output_kband_f=300mm', bbox_inches='tight', dpi=300)


#--------------Extracting a line profile---------------#
width         = 25
x             = np.linspace(0,width*2-1,width*2)

#-----------------k-band-----------------#
#left peak max at 23
xoffset1k     = 47
LinProfleftk  = ImCropk[2,24,xoffset1k-width:xoffset1k+width]

#central peak max at 22/21
xoffset2k     = 160
LinProfcentk  = ImCropk[2,27,xoffset2k-width:xoffset2k+width]

#right peak max at line 26/25
xoffset3k     = 280
LinProfrightk = ImCropk[2,27,xoffset3k-width:xoffset3k+width]

'''
plt.figure()
plt.plot(LinProfleftk, c='b')
plt.plot(LinProfcentk, c='g')
plt.plot(LinProfrightk, c='orange')
'''

#-----------------L-band-----------------#
#left peak max at 23
xoffset1L     = 50
LinProfleftL  = ImCropL[3,25,xoffset1L-width:xoffset1L+width]

#central peak max at 22/21
xoffset2L     = 170
LinProfcentL  = ImCropL[3,25,xoffset2L-width:xoffset2L+width]

#right peak max at line 26/25
xoffset3L     = 285
LinProfrightL = ImCropL[3,25,xoffset3L-width:xoffset3L+width]


plt.figure()
#plt.ylim(0,800)
plt.plot(LinProfleftL, c='b')
plt.plot(LinProfcentL, c='g')
plt.plot(LinProfrightL, c='orange')


#-------------Fitting a Gaussian---------------------#

def gauss(x,  *p):
    a, b, c, d = p
    y = a*np.exp(-np.power((x - b),2.)/(2. * c**2))+ d
    return y

#Initial guess
p_initial1k     = [500, 20, 5, 15]
p_initial2k     = [500, 28, 5, 15]
p_initial3k     = [500, 30, 5, 15]
p_initial1L     = [500, 25, 5, 15]
p_initial2L     = [500, 30, 5, 15]
p_initial3L     = [500, 30, 5, 15]

#Fit
popt1k, pcov1k  = curve_fit(gauss, x, LinProfleftk, p_initial1k)
popt2k, pcov2k  = curve_fit(gauss, x, LinProfcentk, p_initial2k)
popt3k, pcov3k  = curve_fit(gauss, x, LinProfrightk, p_initial3k)
popt1L, pcov1L  = curve_fit(gauss, x, LinProfleftL, p_initial1L)
popt2L, pcov2L  = curve_fit(gauss, x, LinProfcentL, p_initial2L)
popt3L, pcov3L  = curve_fit(gauss, x, LinProfrightL, p_initial3L)

#Generate the best fit y data
y_fit1k         = np.zeros_like(x)
y_fit2k         = np.zeros_like(x)
y_fit3k         = np.zeros_like(x)
y_fit1L         = np.zeros_like(x)
y_fit2L         = np.zeros_like(x)
y_fit3L         = np.zeros_like(x)

for i in range(np.size(y_fit1k)):
    y_fit1k[i] = gauss(x[i], *popt1k)
    y_fit2k[i] = gauss(x[i], *popt2k)
    y_fit3k[i] = gauss(x[i], *popt3k)
    y_fit1L[i] = gauss(x[i], *popt1L)
    y_fit2L[i] = gauss(x[i], *popt2L)
    y_fit3L[i] = gauss(x[i], *popt3L)





MFDPixk         = np.zeros((3))
MFDPixk[0]      = popt1k[2]*4
MFDPixk[1]      = popt2k[2]*4
MFDPixk[2]      = popt3k[2]*4
MFDPixAvek      = np.mean(MFDPixk)
MFDPixErrork    = np.std(MFDPixk)

MFDPixL         = np.zeros((3))
MFDPixL[0]      = popt1L[2]*4
MFDPixL[1]      = popt2L[2]*4
MFDPixL[2]      = popt3L[2]*4
MFDPixAveL      = np.mean(MFDPixL)
MFDPixErrorL    = np.std(MFDPixL)

peakleftk       = popt1k[1]+xoffset1k-width
peakcentk       = popt2k[1]+xoffset2k-width
peakrightk      = popt3k[1]+xoffset3k-width
peakleftL       = popt1L[1]+xoffset1L-width
peakcentL       = popt2L[1]+xoffset2L-width
peakrightL      = popt3L[1]+xoffset3L-width

peakDeltak      = np.zeros((3))
peakDeltak[0]   = peakcentk-peakleftk
peakDeltak[1]   = peakrightk-peakcentk
peakDeltak[2]   = (peakrightk-peakleftk)/2

peakDeltaL      = np.zeros((3))
peakDeltaL[0]   = peakcentL-peakleftL
peakDeltaL[1]   = peakrightL-peakcentL
peakDeltaL[2]   = (peakrightL-peakleftL)/2

coreSpacingPixk = np.mean(peakDeltak)
errorPixk       = np.std(peakDeltak)
coreSpacingPixL = np.mean(peakDeltaL)
errorPixL       = np.std(peakDeltaL)

FibPosXOutBund2  = (np.array((-253.2, -380.8, -510.8, -637.6, -763.0,)))*-1 

peakDeltaMicrons = np.zeros((10))
peakDeltaMicrons[0] = FibPosXOutBund2[1]-FibPosXOutBund2[0]
peakDeltaMicrons[1] = FibPosXOutBund2[2]-FibPosXOutBund2[1]
peakDeltaMicrons[2] = FibPosXOutBund2[3]-FibPosXOutBund2[2]
peakDeltaMicrons[3] = FibPosXOutBund2[4]-FibPosXOutBund2[3]
peakDeltaMicrons[4] = (FibPosXOutBund2[2]-FibPosXOutBund2[0])/2
peakDeltaMicrons[5] = (FibPosXOutBund2[3]-FibPosXOutBund2[1])/2
peakDeltaMicrons[6] = (FibPosXOutBund2[4]-FibPosXOutBund2[2])/2
peakDeltaMicrons[7] = (FibPosXOutBund2[3]-FibPosXOutBund2[0])/3
peakDeltaMicrons[8] = (FibPosXOutBund2[4]-FibPosXOutBund2[1])/3
peakDeltaMicrons[9] = (FibPosXOutBund2[4]-FibPosXOutBund2[0])/4

coreSpacingMicrons  = np.mean(peakDeltaMicrons)
errorMicrons        = np.std(peakDeltaMicrons)

micronsPerPixk      = coreSpacingMicrons/coreSpacingPixk
micronsPerPixErrork = ((errorMicrons/coreSpacingMicrons)+(errorPixk/coreSpacingPixk))*micronsPerPixk
micronsPerPixL      = coreSpacingMicrons/coreSpacingPixL
micronsPerPixErrorL = ((errorMicrons/coreSpacingMicrons)+(errorPixL/coreSpacingPixL))*micronsPerPixL 

MFDmicronsk         = MFDPixAvek*micronsPerPixk
MFDerrork           = ((MFDPixErrork/MFDPixAvek)+(micronsPerPixErrork/micronsPerPixk))*MFDmicronsk
MFDmicronsL         = MFDPixAveL*micronsPerPixL
MFDerrorL           = ((MFDPixErrorL/MFDPixAveL)+(micronsPerPixErrorL/micronsPerPixL))*MFDmicronsL


#Create a plot

fig1 = plt.figure()
plt.plot((x[5:35]-8)/micronsPerPixk,LinProfleftk[5:35], c='b')
plt.plot((x[5:35]-8)/micronsPerPixk,y_fit1k[5:35], c='g')
plt.plot((x[15:45]+10)/micronsPerPixk,LinProfcentk[15:45], c='b')
plt.plot((x[15:45]+10)/micronsPerPixk,y_fit2k[15:45], c='g')
plt.plot((x[15:45]+32)/micronsPerPixk,LinProfrightk[15:45], c='b')
plt.plot((x[15:45]+32)/micronsPerPixk,y_fit3k[15:45], c='g')
plt.title('Line profile of 3 in focus modes in k-band')
plt.ylim(-10,900)
plt.xlim(0,70)
plt.xlabel('Distance (microns)')
plt.ylabel('Counts (ADU)')
#fig1.savefig('Mode_fields_for_the_kband_modes_bundle2.png', bbox_inches='tight', dpi=300)

fig2 = plt.figure()
plt.plot((x[13:43]-11)/micronsPerPixL,LinProfleftL[13:43], c='b')
plt.plot((x[13:43]-11)/micronsPerPixL,y_fit1L[13:43], c='g')
plt.plot((x[12:42]+9)/micronsPerPixL,LinProfcentL[12:42], c='b')
plt.plot((x[12:42]+9)/micronsPerPixL,y_fit2L[12:42], c='g')
plt.plot((x[14:44]+27)/micronsPerPixL,LinProfrightL[14:44], c='b')
plt.plot((x[14:44]+27)/micronsPerPixL,y_fit3L[14:44], c='g')
plt.title('Line profile of 3 in focus modes in L-band')
plt.ylim(-10,800)
plt.xlim(0,70)
plt.xlabel('Distance (microns)')
plt.ylabel('Counts (ADU)')
#fig2.savefig('Mode_fields_for_the_Lband_modes_bundle2.png', bbox_inches='tight', dpi=300)

print(MFDmicronsk)
print(MFDerrork)

print(MFDmicronsL)
print(MFDerrorL)


fig3 = plt.figure()
plt.semilogy((x[5:35]-8)/micronsPerPixk,LinProfleftk[5:35], c='b')
plt.semilogy((x[5:35]-8)/micronsPerPixk,y_fit1k[5:35], c='g')
plt.semilogy((x[15:45]+13)/micronsPerPixk,LinProfcentk[15:45], c='b')
plt.semilogy((x[15:45]+13)/micronsPerPixk,y_fit2k[15:45], c='g')
plt.semilogy((x[15:45]+42)/micronsPerPixk,LinProfrightk[15:45], c='b')
plt.semilogy((x[15:45]+42)/micronsPerPixk,y_fit3k[15:45], c='g')
plt.title('Line profile of 3 in focus modes in k-band')
#plt.ylim(-10,900)
plt.xlim(0,80)
plt.xlabel('Distance (microns)')
plt.ylabel('Counts (ADU)')
#fig3.savefig('Mode_fields_for_the_kband_modes_bundle2_log.png', bbox_inches='tight', dpi=300)

#figure()
#plot(ImCropL[10,23,:])
#print(peakind)

```
